# Mintpy

```python
%%capture --no-display
# %tb
EXAMPLE='''
plot_data.py AntisanaSenDT142/mintpy AntisanaSenAT120/mintpy --plot-type velocity --dem AntisanaSenDT142/dem/Antisana_dem.tif --mask-thresh 0.7 --style scatter --scatter-size 2 --shade-exag 0.01 --vlim -3 3 --ref-lalo -0.44 -78.17
'''
lines = EXAMPLE.splitlines()
os.chdir(os.getenv('SCRATCHDIR'))
for line in lines[1:]:
    line = line.strip()
    cmd=line.strip().split(' ')
    sys.argv = cmd
    plot_data.main(sys.argv)
```
